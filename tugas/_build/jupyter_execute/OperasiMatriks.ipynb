# **Aljabar Matriks**

## **1.Matriks**

**Matriks** Adalah suatu susunan bilangan yang disusun dalam baris dan kolom sehingga membentuk jajaran persegi panjang. Seperti conoh berikut:

$$
A =
\begin{bmatrix}
a1 & a2\\
a3 & a4\\
\end{bmatrix}
$$

## **2.Jenis-jenis Matriks**
Matriks juga memiliki beberapa jenis, seperti:


1. **Matriks Nol**\
Matriks Nol adalah matriks yang elemen penyusunnya adalah nol, contoh:

$$
\begin{bmatrix}
0&0\\
0&0\\
\end{bmatrix}
$$

2. **Matriks Baris**/
Matriks Baris adlah matriks yang elemennya terdiri hanya satu baris, contohnya:

$$
\begin{bmatrix}
5&2&4
\end{bmatrix}
$$

3. **Matriks Kolom**\
Matriks kolom adlah matriks yang elemennya terdiri hanya satu kolom, contohnya:

$$
\begin{bmatrix}
2\\
3\\
\end{bmatrix}
$$

4. **Matriks Tranpose**\
Matriks transpose merupakan matriks yang diperoleh dari memindahkan elemen-elemen baris menjadi elemen pada kolom atau sebaliknya. contoh:

$$
A 3x2 =
\begin{bmatrix}
1 & 2\\
3 & 4\\
5 & 6\\
\end{bmatrix}
>>
At =
\begin{bmatrix}
1 & 3 & 5\\
2 & 4 & 6\\
\end{bmatrix}
$$



## **3.Operasi Matriks (Penjumlahan, Pengurangan, Perkalian, dan Transpose)**

### **1. Penjumlahan dan Pengurangan:**
Operasi matriks yang pertama kita bahas adalah penjumlahan matriks. Operasi matriks adalah penjumlahan 2 matriks yang letak komponennya sama. Penjumlahan matriks tidak ada bedanya dengan penjumlahan pada umumnya tetapi harus memenuhi syarat yaitu jumlah ordo yang sama

Contoh Penjumlahan matriks:

$$
A + B =
\begin{bmatrix}
a1 & a2\\
a3 & a4\\
\end{bmatrix}
+
\begin{bmatrix}
b1 & b2\\
b3 & b4\\
\end {bmatrix}\\
C =
\begin{bmatrix}
a1+b1 & a2+b2\\
a3+b3 & a4+b4\\
\end{bmatrix}
$$

Pengurangan matriks juga sama halnya dengan contoh matriks diatas hanya saja operasinya adalah pengurangan

### **2. Perkalian Matriks**
Perkalian matriks sendiri adalah proses mengalikan setiap elemen baris pada matriks pertama dengan elemen kolom pada matriks kedua. Hasilnya kemudian dijumlahkan untuk memperoleh hasil akhir perkalian matriks. Contoh perkalian pada Matriks dengan ordo 2x2:

$$
AB =
\begin{bmatrix}
1 & 2\\
3 & 4\\
\end{bmatrix}
.
\begin{bmatrix}
5&6\\
7&8
\end{bmatrix}\\
=
\begin{bmatrix}
1(5)+2(7) & 1(6)+2(8)\\
3(5)+4(7) & 3(6)+4(8)
\end{bmatrix}\\
=
\begin{bmatrix}
19 & 22\\
50 & 43
\end{bmatrix}
$$

### **3.Tranpose**
Seperti pada **No. 4** pada **Jenis-jenis Matriks**, Matriks transpose merupakan matriks yang diperoleh dari memindahkan elemen-elemen baris menjadi elemen pada kolom atau sebaliknya.Dengan contoh yang sama:

$$
A 3x2 =
\begin{bmatrix}
1 & 2\\
3 & 4\\
5 & 6\\
\end{bmatrix}
>>
At =
\begin{bmatrix}
1 & 3 & 5\\
2 & 4 & 6\\
\end{bmatrix}
$$


## **Berikut adalah salah satu Program untuk mengalikan Matriks :**

```python
def perkalian_matriks(A, B):
    baris_A = len(A)
    kolom_A = len(A[0])
    baris_B = len(B)
    kolom_B = len(B[0])

    # Membuat matriks hasil dengan ukuran yang sesuai
    result = [[0 for _ in range(kolom_B)] for _ in range(baris_A)]

    # Perkalian matriks
    for i in range(baris_A):
        for j in range(kolom_B):
            for k in range(kolom_A):
                result[i][j] += A[i][k] * B[k][j]

    return result

# Matriks A
A = [[3, 1, 5],
     [3, 5, 4],
     [6, 2, 1]]

# Matriks B
B = [[3, 1, 4],
     [9, 3, 4],
     [3, 6, 2]]

# Mengalikan dua matriks
C = perkalian_matriks(A, B)

# Menampilkan hasil
for row in C:
    print(row)
```
