# **Tranformasi Matriks**

Transformasi matriks adalah operasi yang mengubah satu set data atau vektor menjadi set data atau vektor lain melalui penggunaan matriks. Dalam konteks aljabar linier, transformasi ini sering digunakan untuk melakukan berbagai operasi geometris seperti rotasi, skala, dan translasi pada vektor-vektor dalam ruang Euclidean.

## **Jenis-Jenis Transformasi Matriks dan Rumusnya**

### **1. Transformasi Identitas**

Transformasi identitas adalah transformasi yang tidak mengubah vektor. Matriks identitas biasanya dinotasikan dengan I.

Rumus:

$$
[ I = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} ]
$$

### **2. Transformasi Skala (Scaling)**

Transformasi skala memperbesar atau memperkecil vektor dengan faktor tertentu.

Rumus:

$$
[ S = \begin{pmatrix} s_x & 0 \\ 0 & s_y \end{pmatrix} ]
$$

Di mana (s_x) dan (s_y) adalah faktor skala sepanjang sumbu-x dan sumbu-y.

### **3. Transformasi Rotasi (Rotation)**
Transformasi rotasi memutar vektor sebesar sudut tertentu terhadap titik asal.

Rumus:

$$
[ R(\theta) = \begin{pmatrix} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{pmatrix} ]
$$

Di mana (\theta) adalah sudut rotasi.

### **4. Transformasi Translasi (Translation)**
Translasi menggeser vektor dari satu posisi ke posisi lain. Translasi biasanya tidak bisa direpresentasikan dengan matriks 2x2 biasa, tetapi bisa dengan menggunakan matriks homogen.

Rumus (Menggunakan Matriks Homogen 3x3):

$$
[ T = \begin{pmatrix} 1 & 0 & t_x \\ 0 & 1 & t_y \\ 0 & 0 & 1 \end{pmatrix} ]
$$

Di mana (t_x) dan (t_y) adalah jarak translasi sepanjang sumbu-x dan sumbu-y.

### **5. Transformasi Refleksi (Reflection)**
Refleksi mencerminkan vektor terhadap sumbu tertentu.

Rumus (Refleksi terhadap sumbu-x):

$$
[ F_x = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} ]
$$


Rumus (Refleksi terhadap sumbu-y):

$$
[ F_y = \begin{pmatrix} -1 & 0 \\ 0 & 1 \end{pmatrix} ]
$$

### **6. Transformasi Geser (Shear)**
Transformasi geser mengubah bentuk vektor dengan menggesernya sepanjang sumbu-x atau sumbu-y.

Rumus (Shear horizontal):

$$
[ Sh_x = \begin{pmatrix} 1 & k_x \\ 0 & 1 \end{pmatrix} ]
$$

Rumus (Shear vertikal):

$$
[ Sh_y = \begin{pmatrix} 1 & 0 \\ k_y & 1 \end{pmatrix} ]
$$

Di mana (k_x) dan (k_y) adalah faktor geser horizontal dan vertikal.

### **Contoh Implementasi di Python (NumPy)**

```python
import numpy as np

# Contoh vektor
v = np.array([1, 1])

# Matriks Rotasi 90 derajat
theta = np.radians(90)
R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

# Menerapkan transformasi rotasi
v_rotated = R @ v

print("Vektor asli:", v)
print("Vektor setelah rotasi 90 derajat:", v_rotated)
```

```python
#Rotasi Transformasi
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menggambar sumbu dan grid
def draw_axes_and_grid(ax, lim=10):
    ax.axhline(0, color='grey', linewidth=0.5)
    ax.axvline(0, color='grey', linewidth=0.5)
    ax.set_xlim(-lim, lim)
    ax.set_ylim(-lim, lim)
    ax.set_aspect('equal', adjustable='box')
    ax.grid(which='both', linestyle='--', linewidth=0.5)

# Fungsi untuk melakukan transformasi matriks
def transform_and_plot(matrix, points, ax, color='blue', label='Transformed'):
    transformed_points = matrix @ points
    ax.scatter(transformed_points[0, :], transformed_points[1, :], color=color, label=label)
    return transformed_points

# Definisi titik-titik (himpunan titik)
points = np.array([
    [1, 2, 3, 4, 5],  # X-koordinat
    [1, 2, 3, 4, 5]   # Y-koordinat
])

# Sudut rotasi dalam derajat
theta_degrees = 45
theta = np.radians(theta_degrees)

# Matriks transformasi untuk rotasi
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])

# Plot hasil transformasi
fig, ax = plt.subplots()
draw_axes_and_grid(ax)
ax.scatter(points[0, :], points[1, :], color='red', label='Ori')
transformed_points = transform_and_plot(rotation_matrix, points, ax, color='blue', label=f'Rotasi {theta_degrees}')

ax.legend()
plt.show()
```

```python
#refleksi transformasi
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menggambar sumbu dan grid
def draw_axes_and_grid(ax, lim=10):
    ax.axhline(0, color='grey', linewidth=0.5)
    ax.axvline(0, color='grey', linewidth=0.5)
    ax.set_xlim(-lim, lim)
    ax.set_ylim(-lim, lim)
    ax.set_aspect('equal', adjustable='box')
    ax.grid(which='both', linestyle='--', linewidth=0.5)

# Fungsi untuk melakukan transformasi matriks
def transform_and_plot(matrix, points, ax, color='blue', label='Transfor'):
    transformed_points = matrix @ points
    ax.scatter(transformed_points[0, :], transformed_points[1, :], color=color, label=label)
    return transformed_points

# Definisi titik-titik (himpunan titik)
points = np.array([
    [1, 2, 3, 4, 5],  # X-koordinat
    [1, 2, 3, 4, 5]   # Y-koordinat
])

# Matriks transformasi untuk refleksi terhadap sumbu x
reflection_matrix = np.array([
    [1, 0],
    [0, -1]
])

# Plot hasil transformasi
fig, ax = plt.subplots()
draw_axes_and_grid(ax)
ax.scatter(points[0, :], points[1, :], color='red', label='Ori')
transformed_points = transform_and_plot(reflection_matrix, points, ax, color='blue', label='sumbu X')

# Ulangi dengan matriks transformasi lain untuk refleksi terhadap sumbu y dan garis y = x
reflection_matrix_y = np.array([
    [-1, 0],
    [0, 1]
])
transform_and_plot(reflection_matrix_y, points, ax, color='green', label='sumbu Y')

reflection_matrix_yx = np.array([
    [0, 1],
    [1, 0]
])
transform_and_plot(reflection_matrix_yx, points, ax, color='purple', label='sumbu Y=X')

ax.legend()
plt.show()
```
