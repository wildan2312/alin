## **Matriks Invers**

Invers matriks adalah kebalikan (invers) dari sebuah matriks yang apabila matriks tersebut dikalikan dengan inversnya, akan menjadi matriks identitas. Invers matriks dilambangkan dengan A-1. Suatu matriks dikatakan memiliki invers jika determinan dari matriks tersebut tidak sama dengan nol.

### **Sifat-sifat Matriks Invers**



1.   Matriks Invers hanya ada pada matrik yang memiliki jumlah baris dan kolom yang sama
2.    Invers matriks hanya ada untuk matriks nonsingular, yaitu matriks yang memiliki determinan tidak sama dengan nol.
3. Invers matriks adalah tunggal, yaitu tidak ada lebih dari satu matriks yang dapat menjadi invers dari suatu matriks. Jika B dan C keduanya adalah invers dari A, maka B=C.




### **Mencari Invers Dengan Eleminasi Gauss Jordan**

Eliminasi Gauss-Jordan adalah sebuah metode matematis untuk mencari invers dari sebuah matriks. Langkah-langkahnya mirip dengan eliminasi Gauss biasa, tetapi setelah matriks berada dalam bentuk eselon, langkah selanjutnya adalah mengubahnya menjadi bentuk eselon tereduksi. Berikut adalah langkah-langkahnya:

Misalkan kita memiliki sebuah matriks A yang ingin kita cari inversnya.

1. Bentuk matriks tambahan dengan matriks identitas yang sama ukurannya. Misalnya, jika A adalah matriks n x n, maka matriks tambahan ini akan menjadi matriks identitas I yang juga n x n, seperti ini:

$$
\begin{bmatrix}
A \mid I
\end{bmatrix}
$$

2. Terapkan metode eliminasi Gauss pada matriks tambahan sehingga matriks A menjadi bentuk eselon. Ini akan menghasilkan matriks tambahan dalam bentuk:

$$
\begin{bmatrix}
E \mid B
\end{bmatrix}
$$

Di mana E adalah matriks yang menggambarkan langkah-langkah eliminasi Gauss yang diterapkan pada matriks A , dan B adalah matriks yang sama dengan invers dari A.

3. Terapkan operasi baris pada E | B sehingga matriks E menjadi matriks identitas. Hasil akhirnya akan menjadi [ I | A^{-1} ], di mana A^{-1} adalah invers dari A.

Berikut adalah contoh langkah-langkahnya:

Misalkan kita memiliki matriks A:

$$
A = \begin{bmatrix}
2 & 1 \\
1 & -3
\end{bmatrix}
$$

Kita ingin mencari invers dari A.

Langkah 1: Bentuk matriks tambahan

$$
\begin{bmatrix}
2 & 1 & | & 1 & 0 \\
1 & -3 & | & 0 & 1
\end{bmatrix}
$$

Langkah 2: Terapkan eliminasi Gauss-Jordan

$$
\begin{align*}
R_2 & = R_2 - \frac{1}{2}R_1 \\
& \Rightarrow
\begin{bmatrix}
2 & 1 & | & 1 & 0 \\
0 & -\frac{7}{2} & | & -\frac{1}{2} & 1
\end{bmatrix}
\end{align*}
$$

$$
\begin{align*}
R_1 & = R_1 - \frac{2}{7}R_2 \\
& \Rightarrow
\begin{bmatrix}
2 & 0 & | & \frac{3}{7} & -\frac{2}{7} \\
0 & -\frac{7}{2} & | & -\frac{1}{2} & 1
\end{bmatrix}
\end{align*}
$$

$$
\begin{align*}
R_2 & = -\frac{2}{7}R_2 \\
& \Rightarrow
\begin{bmatrix}
2 & 0 & | & \frac{3}{7} & -\frac{2}{7} \\
0 & 1 & | & \frac{1}{7} & -\frac{2}{7}
\end{bmatrix}
\end{align*}
$$

$$
\begin{align*}
R_1 & = R_1 + R_2 \\
& \Rightarrow
\begin{bmatrix}
2 & 0 & | & 1 & -\frac{4}{7} \\
0 & 1 & | & \frac{1}{7} & -\frac{2}{7}
\end{bmatrix}
\end{align*}
$$

Langkah 3: Mengonversi [ E|B ] menjadi [ I|A^{-1} ]

Matriks A^{-1} adalah:

$$
A^{-1} = \begin{bmatrix}
1 & -\frac{4}{7} \\
\frac{1}{7} & -\frac{2}{7}
\end{bmatrix}
$$

Itulah cara menghitung invers matriks menggunakan metode eliminasi Gauss-Jordan. Langkah-langkahnya bisa lebih rumit tergantung pada ukuran dan sifat matriks tersebut.

### **Solusi Persamaan Linier Dengan Matriks Invers**

Soal Persamaan Linier 6 Variabel:

Berikut adalah sebuah contoh sistem persamaan linear dengan 6 variabel:

$$

\begin{cases}
2x_1 + 3x_2 - x_3 + 4x_4 + x_5 + 3x_6 = 12 \\
4x_1 + 2x_2 + 5x_3 - 3x_4 - 2x_5 + x_6 = 4 \\
3x_1 + x_2 + 2x_3 + 4x_4 - x_5 + 2x_6 = 8 \\
x_1 - 3x_2 + 4x_3 - 2x_4 + 3x_5 - 2x_6 = -3 \\
2x_1 + 4x_2 + x_3 + 5x_4 - 4x_5 + 3x_6 = 15 \\
x_1 + 2x_2 + 3x_3 - x_4 + 2x_5 + 4x_6 = 10 \\
\end{cases}

$$


```python
import numpy as np

# Matriks koefisien
A = np.array([[2, 3, -1, 4, 1, 3],
              [4, 2, 5, -3, -2, 1],
              [3, 1, 2, 4, -1, 2],
              [1, -3, 4, -2, 3, -2],
              [2, 4, 1, 5, -4, 3],
              [1, 2, 3, -1, 2, 4]])

# Vektor hasil
B = np.array([12, 4, 8, -3, 15, 10])

# Mencari invers matriks A
A_inv = np.linalg.inv(A)

# Mencari solusi
X = np.dot(A_inv, B)

print("Solusi:")
for i in range(len(X)):
    print(f"x{i+1} =", X[i])
```
